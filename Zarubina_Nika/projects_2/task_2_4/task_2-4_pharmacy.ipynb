```python
kapsula = int(input("Общее количество произведенных капсул: "))
mesto = int(input("Количество капсул в одной упаковке: "))
full = kapsula // mesto
ostatok = kapsula % mesto
print("\n--- Отчет фасовочного цеха ---")
print(f"Полных упаковок:\t{full}")
print(f"Остаток капсул:\t\t{ostatok}")
```
