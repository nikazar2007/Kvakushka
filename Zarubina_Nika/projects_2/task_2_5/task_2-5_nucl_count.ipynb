```python
print('=== Анализ последовательности ДНК ===')
dna = input("Введите последовательность ДНК:").upper()
print("\nПоследовательность в верхнем регистре:", dna)
print("\nПодсчёт нуклеотидов:")
print("A:", dna.count("A"))
print("G:", dna.count("G"))
print("T:", dna.count("T"))
print("C:", dna.count("C"))
l_dna = dna.count("A") + dna.count("G") + dna.count("T") + dna.count("C") 
print("\nОбщая длина:", l_dna)
Apr = dna.count("A") / l_dna * 100
Gpr = dna.count("G") / l_dna * 100
Tpr = dna.count("T") / l_dna * 100
Cpr = dna.count("C") / l_dna * 100
print("\nПроцентное содержание каждого нуклеотидa: ")
print(f"A: {Apr:.2f}%\nG: {Gpr:.2f}%\nT: {Tpr:.2f}%\nC: {Cpr:.2f}%")
```
