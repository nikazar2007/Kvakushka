```python
print("\nТри дороги: налево, направо и прямо.")
print("Куда направишься? Прямо, налево, направо?")
choice = input().strip().lower()
if choice == "прямо" or choice == "1":
    print("\nПоехал прямо — умер в трясине :(")
elif choice == "налево" or choice == "2":
    print("\nПовернул налево — обрёл много монет, но душевного спокойствия у тебя не будет :|")
elif choice == "направо" or choice == "3":
    print("\nСвернул направо — встретил женщину, у тебя всё будет хорошо :")
else:
    print("\nНеведомая дорога... Витязь задумался и остался стоять у камня.")
```
